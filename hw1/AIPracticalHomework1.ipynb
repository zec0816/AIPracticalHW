# **Python编程作业第一部分：学习NumPy库的常用函数**

一、数组创建与使用

```python
arr1 = np.arange(10)
print("1D array:", arr1)

arr2 = arr1 * 2
print("Multiplied by 2:", arr2)
print("Shape:", arr2.shape) #1D, 10 elements

arr3 = np.arange(1, 10).reshape(3, 3)
print("3x3 array:\n", arr3)
print("Shape:", arr3.shape) #2D, 3 elements per row, 3 rows total

print("Sum:", np.sum(arr3))
print("Mean:", np.mean(arr3))
print("Max:", np.max(arr3))
print("Min:", np.min(arr3))
```

二、数组切片与变形

```python
arr = np.arange(16).reshape(4, 4)
print("Original 4x4 array:\n", arr)

element = arr[1, 2]
print("Element of R2C3:", element)

sub_arr = arr[0:2, 1:3]
print("Subarray(R1,2 & C2,3):\n", sub_arr)

reshaped_arr = arr.reshape(2, 8)
print("Reshaped to 2x8 array:\n", reshaped_arr)
```

三、数组的运算

```python
arr1 = np.random.permutation(np.arange(1, 10)).reshape(3, 3)
arr2 = np.random.permutation(np.arange(10, 19)).reshape(3, 3)

print("Array 1:\n", arr1)
print("Array 2:\n", arr2)

add_result = arr1 + arr2
sub_result = arr1 - arr2
mult_result = np.dot(arr1, arr2)

print("\nAddition result:\n", add_result)
print("\nSubtraction result:\n", sub_result)
print("\nMultiplication result:\n", mult_result)

arr1_flat = arr1.reshape(-1)
arr2_flat = arr2.reshape(-1)

print("\nFlatten array 1:", arr1_flat)
print("Flatten array 2:", arr2_flat)

dot_result = np.dot(arr1_flat, arr2_flat)
print("\nDot result:", dot_result)
```

四、统计函数的使用

```python
arr = np.random.randint(0, 100, size=20)
print("Random integer array:\n", arr)

greater_than_fifty = arr[arr > 50]
print("Elements >50:\n", greater_than_fifty)

var = np.var(arr)
std = np.std(arr)
print("Variance:", var)
print("Standard deviation:", std)
```

五、数组的合并与分割

```python
arr1 = np.array([0, 1, 2])
arr2 = np.array([3, 4, 5])
print("Array 1:", arr1)
print("Array 2:", arr2)

hori_concat = np.hstack((arr1, arr2))
print("\nHorizontal concat result:\n", hori_concat)

ver_concat = np.vstack((arr1, arr2))
print("\nVertical concat result:\n", ver_concat)

a_col = arr1.reshape(3, 1)
b_col = arr2.reshape(3, 1)
h_concat_cols = np.hstack((a_col, b_col))
print("\nHorizontal concat of column vectors result:\n", h_concat_cols)

split_result = np.split(hori_concat, 3)
print("\nSplit result：")
print("、".join([str(arr) for arr in split_result]))
```

# **Python编程作业第二部分：基本图像处理**

一、颜色转换与图像保存

```python
img = cv2.imread('hw1.png')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

print("Grayscale image:")
cv2_imshow(gray)
cv2.imwrite('gray.jpg', gray)
```

二、图像的基本处理： 缩放、旋转、平移

```python
img = cv2.imread('hw1.png')

# Image scaling
height, width = img.shape[:2]
scaled_img = cv2.resize(img, (width // 2, height // 2))

print(f"Original size: {width}x{height}")
print(f"Scaled size: {scaled_img.shape[1]}x{scaled_img.shape[0]}")

print("\nScaled image:")
cv2_imshow(scaled_img)
cv2.imwrite('scaled.jpg', scaled_img)
print()

# Image rotation
(h, w) = img.shape[:2]
center = (w // 2, h // 2)
rotation_matrix = cv2.getRotationMatrix2D(center, 30, 1.0)
rotated_img = cv2.warpAffine(img, rotation_matrix, (w, h))

print("Rotated image:")
cv2_imshow(rotated_img)
cv2.imwrite('rotated.jpg', rotated_img)
print()

# Image translation
M = np.float32([[1, 0, 50], [0, 1, 100]])
translated_img = cv2.warpAffine(img, M, (w, h))

print("Translated image:")
cv2_imshow(translated_img)
cv2.imwrite('translated.jpg', translated_img)
```
